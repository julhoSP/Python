```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([100, 200, 300, 500, 400, 150])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, ls ='--')
#plt.scatter(salario, marker = "*", color='red')
plt.show()
```

```python
salarios_marcus = np.array([100,200,400,300])
salarios_gileno = np.array([50,300,500,450])
```

```python
plt.plot(salarios_marcus, c='Black', ls='--', marker='s', ms=8,label='Salarios Marcus')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8,label='Salarios Gileno')
plt.legend()
plt.show()
```

```python
salarios_pedro = [200, 150, 300, 700]
plt.plot(salarios_marcus, c='Black', ls='--', marker='s', ms=8,label='Salarios Marcus')
plt.plot(salarios_gileno, c='Red', ls='--', marker='^', ms=8,label='Salarios Gileno')
plt.plot(salarios_pedro, c='Blue', ls='--', marker='o', ms=8,label='Salarios Pedro')
plt.title('Salarios dos Funcionarios')
plt.ylabel('Valor em Reais')
plt.xlabel('Meses')
plt.legend()
plt.show()
```

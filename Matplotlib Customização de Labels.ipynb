```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('reported.csv')
```

```python
df.head()
```

```python
df.fillna(0,inplace= True)
```

```python
df.head()
```

```python
plt.plot(df['Year'],df['crimes.total'],'-r',)
plt.plot(df['Year'],df['crimes.person'],'-b')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'],df['crimes.total'],'-r')
ax.plot(df['Year'],df['crimes.person'],'-b')
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'],df['crimes.total'],'-r')
ax.plot(df['Year'],df['crimes.person'],'-b')
ax.legend()
ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')
ax.set_title('Crimes: Total x Pessoa')
```

```python
fig, ax = plt.subplots()
ax.plot(df['Year'],df['crimes.total'],'-r')
ax.plot(df['Year'],df['crimes.person'],'-b')
ax.legend()
ax.set_ylabel('Quantidade')
ax.set_xlabel('Ano')
ax.set_title('Crimes: Total x Pessoa')
ax.set_xlim([df['Year'].min()-5, df['Year'].max()+5])
ax.plot()
```

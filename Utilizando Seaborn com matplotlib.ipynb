```python
iris = pydataset.data('iris')
```

```python
iris.head()
```

```python
plt.scatter(iris['Sepal.Length'],iris['Sepal.Width'], sizes=50 * iris['Petal.Length'])
```

```python
def specie_color(x):
    if x == 'setosa':
        return 0
    elif x == 'versicolor':
        return 1
    return 2
```

```python
iris['SpeciesNumber'] = iris['Species'].apply(specie_color)
```

```python
iris.head()
```

```python
plt.scatter(
    iris['Sepal.Length'],
    iris['Sepal.Width'], 
    sizes= 20*iris['Petal.Length'],
    c= iris['SpeciesNumber'],
    cmap= 'viridis',
    alpha=0.4
)
plt.legend()
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.genfromtxt('irisdata.csv', delimiter=',',usecols=(0,1,2,3))
data
```

```python
plt.plot(data[:50,0],c='Red', ls=':', marker='s', label='Comp. Iris Setosa')
plt.plot(data[50:100,0],c='Blue', ls=':', marker='o',label='Comp. Iris Versicolor')
plt.legend()
plt.show()
```

```python
plt.plot(data[:50,0],c='Red', ls=':', marker='s', label='Comp. Iris Setosa')
plt.plot(data[50:100,0],c='Blue', ls=':', marker='o',label='Comp. Iris Versicolor')
plt.plot(data[100:150,0],c='Green', ls=':', marker='^',label='Comp. Iris Virginica')
plt.title('Comprimento de Sepala')
plt.ylabel('Comprimento em cm')
plt.legend()
plt.show()
```

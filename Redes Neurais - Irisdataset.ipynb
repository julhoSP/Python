```python
'''
Iris-setosa: 0
Iris-versicolor: 1
Iris-virginica: 2
'''
import numpy as np
from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.supervised import BackpropTrainer
```

```python
entrada = np.genfromtxt('iris.data',delimiter=',', usecols=(0,1,2,3))
saida = np.genfromtxt('iris.data',delimiter=',', usecols=(4))
```

```python
#Utilizar 70 para treino sendo 35 de cada classe os 25 serão teste
entradas_treino = np.concatenate((entrada[:35],entrada[50:85],entrada[100:135]))
saidas_treino = np.concatenate((saida[:35],saida[50:85],saida[100:135]))
```

```python
#Pegar 45 amostras para teste, 15 de cada planta
entradas_teste = np.concatenate((entrada[35:50],entrada[85:100],entrada[135:]))
saidas_teste = np.concatenate((saida[35:50],saida[85:100],saida[135:]))
```

```python
#Criação dataset de treinamento
treinamento = SupervisedDataSet(4,1)
# Inserindo dados de treinamento já previamente separados
for i in range(len(entradas_treino)):
    treinamento.addSample(entradas_treino[i],saidas_treino[i])
```

```python
#Construindo a rede neurais 1 -entradas, neuronios intermediarios, saidas
rede = buildNetwork(treinamento.indim, 2, treinamento.outdim, bias=True)
#Criando treinamento, passa a rede, dataset de treinamento, taxa de aprendizado e momento
trainer = BackpropTrainer(rede, treinamento, learningrate=0.01, momentum=0.3)
```

```python
#Treinando modelo
for epoca in range(1000):
    trainer.train()
    
#Dataset de teste
teste = SupervisedDataSet(4, 1)
#Inserindo dados
for i in range(len(entradas_teste)):
    teste.addSample(entradas_teste[i],saidas_teste[i])
trainer.testOnData(teste, verbose=True)
```

```python
#variando o momentum
trainer = BackpropTrainer(rede, treinamento, learningrate=0.1, momentum=0.9)
```

```python
#Treinando modelo
for epoca in range(1000):
    trainer.train()
    
#Dataset de teste
teste = SupervisedDataSet(4, 1)
#Inserindo dados
for i in range(len(entradas_teste)):
    teste.addSample(entradas_teste[i],saidas_teste[i])
trainer.testOnData(teste, verbose=True)
```

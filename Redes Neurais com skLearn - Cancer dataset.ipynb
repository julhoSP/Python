```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
#carregando dataset
cancer = load_breast_cancer()
#Conhecendo dataset
print(cancer.feature_names)
print(cancer.target_names)
print(cancer.DESCR)
```

```python
x = cancer.data
y = cancer.target
```

```python
#valor padrão é 0.25
xtrain, xtest, ytrain, ytest = train_test_split(x,y,train_size= 0.75)
```

```python
#Normalizando o dados com o StandardScaler
scaler = StandardScaler()
# Treinando o normalizador
scaler.fit(xtrain)
# Passando dados normalizados para o dataset de treino
xtrain = scaler.transform(xtrain)
# Passando dados normalizados para o dataset de test
xtest = scaler.transform(xtest)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30))
mlp.fit(xtrain,ytrain)
```

```python
mlp.score(xtest,ytest)
saida = mlp.predict(xtest)
```

```python
print(confusion_matrix(ytest, saida))
print(classification_report(ytest,saida))
```
